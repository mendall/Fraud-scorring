```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import re
%matplotlib inline
from sklearn import linear_model, datasets, metrics
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import numpy as np
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data'
logistic = pd.read_csv(url, header= None,na_values='?')
```

```python
logistic.describe()
```

```python

logistic = logistic[[15,14,2]]
```

```python
logistic.columns = [i for i in range(0, 3)] 
logistic = logistic.replace(['+','-'],[1,-1])
print(logistic)
```

```python
logistic.describe()
```

```python
logistic = logistic.fillna(logistic.median(axis=0), axis=0)
logistic.count(axis=0)
```

```python
logreg = linear_model.LogisticRegression(C=1e5, solver='lbfgs')
```

```python
regr = logreg.fit(logistic[[1, 2]],logistic[0])
```

```python
X=np.array(logistic[[1, 2]])
Y=np.array(logistic[0])

h = 0.02  

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(10, 8))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```

```python
metrics.accuracy_score(logistic[0],regr.predict(logistic[[1,2]]))
```

```python
0.0966183574879227
```

```python
metrics.roc_auc_score(logistic[0],regr.predict(logistic[[1,2]]))
```

```python
fpr, tpr, thresholds = roc_curve(logistic[0], regr.predict_proba(logistic[[1 , 2]].values)[:, 1])
roc_auc = auc(fpr, tpr)
```

```python
plt.plot(fpr, tpr, label='ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate or (1 - Specifity)')
plt.ylabel('True Positive Rate or (Sensitivity)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
```
